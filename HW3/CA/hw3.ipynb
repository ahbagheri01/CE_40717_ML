<font size =4>
    <br><br>
    the equation of closed form of Linear regression is:
</font>
$$
W = (\Phi^T \Phi+\lambda I)^{-1} \Phi^T Y
$$
$$
pred = \Phi W
$$

<font size =3>
in the cell below we define a regression model and two function to use during the part 3 4 5 6 
</font>

```python
class LinearRegressionModel:
    def __init__(self,phi):
        self.w = 0
        self.phi = phi
    def fit(self,X,y,regularization):
        """
        get X and y train and learn the parameter by the equation.
    
        """
        X = X.to_numpy()
        n,m = X.shape
        y = y.to_numpy().reshape(-1,1)
        X = self.phi(X)
        n,m = X.shape
        C = X.T@X + regularization*np.identity(m)
        Cinv = np.linalg.pinv(C)
        w = Cinv@X.T
        w = w@y
        self.w = w.reshape(m,1)
    def evaluate(self,X,y):
        """
        get X and y and calculate error.
        """
        pred = self.transform(X)
        y = y.to_numpy().reshape(-1,1)
        assert len(pred) == len(y) , "length of predicted and target does not match";
        dif = pred - y
        N,_ = y.shape
        error = dif.T@dif
        return float(error/N)
    def transform(self,X):
        """
        get X and calculate Phi(X)W as predicted values.
    
        """
        X = X.to_numpy()
        X = self.phi(X)
        pred = X@self.w
        return pred
    def get_param(self):
        return self.w.reshape(-1)
def show_data(x,y,pred,title_type,x_axis,save = None):
    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(x, y, c="red", label="mainVal")
    ax.scatter(x, pred, c="blue", label="predictedVal")
    ax.legend()
    plt.xlabel(x_axis)
    plt.ylabel('MEDV') 
    plt.title(x_axis+"-MEDV plot for "+title_type)
    if save!= None:
        plt.savefig(save)
    plt.show()
def show_analysis(model,regularization,features,show_plot=True,show_info=True,train=True,test=True,whole=True):
    if show_info:
        print("model info : regularization = ",regularization);
        print("training set error = ",model.evaluate(X_train,y_train));
        print("test set error = ",model.evaluate(X_test,y_test));
        print("whole datset error = ",model.evaluate(data_set,df["MEDV"]));
        print("the W is :");
        print(model.get_param().reshape(1,-1));
        print("w size(norm2) is :",np.linalg.norm(model.get_param()))
    if show_plot:
        pred_test = model.transform(X_test);
        pred_train = model.transform(X_train);
        pred_all = model.transform(data_set)
        for feature in features:
            if train:
                show_data(X_train[feature].to_numpy(),y_train,pred_train,"trainingSet",feature)
            if test:
                show_data(X_test[feature].to_numpy(),y_test,pred_test,"testSet",feature)
            if whole:
                show_data(data_set[feature].to_numpy(),df["MEDV"].to_numpy(),pred_all,"whole data_set(including test and train)",feature)
```

```python
basic_finctions = X_train.sample(n=10).to_numpy()
def gussian_randomly(X,s,basic_finctions):
    gussian = np.exp(-np.true_divide(np.linalg.norm((X-basic_finctions[0]),axis = 1),2*s*s))
    for index in range(1,10):
        gussian = np.c_[gussian,np.exp(-np.true_divide(np.linalg.norm((X-basic_finctions[index]),axis = 1)**2,2*s*s))]
    result = np.c_[np.ones(X.shape[0]),X,gussian]
    return result
```

```python
train_loss = []
for i in range(1,25):
    phi_3 = lambda X : gussian_randomly(X,i,basic_finctions)
    model3 = LinearRegressionModel(phi_3)
    regularization = 0
    model3.fit(X_train,y_train,regularization)
    train_loss.append(model3.evaluate(X_train,y_train))
fig = plt.figure(figsize=(20,10))
plt.plot(range(1,25),train_loss)
plt.title("training error - S")
plt.show()
```
