# preprocess

import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix
import seaborn as sn
from sklearn import tree 
import scipy.stats as stats
from matplotlib.colors import ListedColormap
```

import dataset

```python
data = pd.read_csv("data.csv",index_col=False)
data = data.drop(["Unnamed: 0"], axis=1)
data
```

plot dataset

```python
def show_data(data,title):
    fig  = plt.figure(figsize=(15,15))
    sns.scatterplot(data=data, x="X1", y="X2", hue="Y")
    plt.title(title)
    plt.show()
show_data(data,"data distribution") 
```

split dataset

```python
X = data[["X1","X2"]]
y = data[["Y"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# model

here is an algorithm we used.

image.png

```python
import math
class adaboost:
    def __init__(self):
        self.trees = []
        self.alphas = []
    def fit(self,N,X,y):
        self.alphas = np.ones(N)
        self.trees = [tree.DecisionTreeClassifier(max_depth =1,criterion ="gini",random_state = 1) for _ in range(N)]
        prob = np.ones(len(X))/len(X)
        for t in range(N):
            self.trees[t].fit(X,y,sample_weight = prob)
            pred = self.trees[t].predict(X).reshape(-1)
            epsilon = sum((pred!=y.to_numpy().reshape(-1))*prob)/sum(prob)
            self.alphas[t] = np.log((1-epsilon)/epsilon)
            prob = self.update(prob,y.to_numpy().reshape(-1),pred.reshape(-1),self.alphas[t])
    def predict(self,X):
        pred = np.array([self.trees[i].predict(X) *self.alphas[i]  for i in range(len(self.trees))])
        pred = np.sum(pred,axis = 0)
        return np.sign(pred)
    def update(self,weights, target, prediction, alpha_t):
        prob =  [w * (math.e ** (alpha_t * int(p != t))) for (w, p, t) in zip(weights, prediction, target)]
        return np.array(prob)
        
```

```python
ada= adaboost()
N = 16
ada.fit(N,X_train,y_train)
```

# Accuracy and f1 score

```python
def visualize_result(y_true,y_pred):
  cm = confusion_matrix(y_true, y_pred)
  df_cm = pd.DataFrame(cm, range(2), range(2))
  plt.figure(figsize=(10,7))
  sn.set(font_scale=1) 
  sn.heatmap(df_cm, annot=True, annot_kws={"size": 20},fmt="d",cmap="YlGnBu") # font size
  from sklearn.metrics import classification_report
  from sklearn.metrics import f1_score,recall_score,precision_score
  target_names = ['class 0', 'class 1']
  print("f1_score "+" is :{}%".format(f1_score(y_true=y_true , y_pred= y_pred)))
  print("recall_score "+" is :{}%".format(recall_score(y_true=y_true , y_pred= y_pred)))
  print("precision_score "+" is :{}%".format(precision_score(y_true=y_true , y_pred= y_pred)))
  print(classification_report(y_true, y_pred, target_names=target_names))
```

```python
visualize_result(y_test,ada.predict(X_test))
```

# compare with just one tree classifier
maximum depth 1

```python
clf = tree.DecisionTreeClassifier(max_depth =1,criterion ="gini",random_state = 1)
clf.fit(X_train,y_train)
visualize_result(y_test,clf.predict(X_test))
```
