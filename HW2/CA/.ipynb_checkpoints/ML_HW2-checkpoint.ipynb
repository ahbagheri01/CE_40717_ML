```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

r1 = np.random.randn(50)/3
r2 = np.random.randn(50)/3
scatter_hue = np.random.choice([True, False], size=50)
r1 += np.ones(50) * ~scatter_hue * 0.5
axes[0][0].set_title('Scatter plot')
sns.scatterplot(x=r1, y=r2, hue=scatter_hue, ax=axes[0][0])

dist_smaples = list(np.random.normal(loc=0.0, scale=1, size=20)) + [8,8,-32]
axes[0][1].set_title('Histogram')
sns.distplot(dist_smaples, bins=20, hist=True, ax=axes[0][1])

pie_values = [10, 30, 2, 45]
axes[1][0].set_title('Pie Chart')
axes[1][0].pie(sorted(pie_values), labels=sorted(pie_values))

box_values = np.concatenate((2 + np.random.randn(20), -5 + 2 * np.random.randn(5)))
axes[1][1].set_title('Box Plot')
sns.boxplot(box_values, ax=axes[1][1]);
```

```python
fig, axes = plt.subplots(9,2, figsize=(15, 60))
axes = axes.flatten()
for i,col in enumerate(numeric_cols):
    sns.boxplot(data=credit_cards,x = str(col),ax=axes[2*i])
    sns.distplot(credit_cards[col], hist=True, kde=True, 
             bins=int(20), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4},ax = axes[2*i+1])
plt.show()
```

```python
def IQR_method(df):
    cols = list((numeric_cols.columns))
    cols.remove("reports")
    cols.remove("majorcards")
    df.where(df[cols] >= 0  ,other=np.nan,inplace=True)
    q1 = df[cols].quantile([0.25]).to_numpy()
    q3 = df[cols].quantile([0.75]).to_numpy()
    lower = pd.DataFrame(data = (2.5*q1 - 1.5*q3) , index = [0], columns = cols) 
    upper = pd.DataFrame(data = (2.5*q3 - 1.5*q1) , index = [0], columns = cols)
    lower = pd.concat([lower]*index_size, ignore_index=True)
    upper = pd.concat([upper]*index_size, ignore_index=True)
    df.where((df[cols] >= lower[cols])[cols] ,other=np.nan,inplace=True)
    df.where((df[cols] <= upper[cols])[cols] ,other=np.nan,inplace=True)
    return df
credit_cards = IQR_method(credit_cards)
credit_cards.head()
```

```python
#helper function
def pdf_hist(df,col,color,title,label = None,show = True,bins = 15,hist = True):
    if show:
        fig = plt.figure(figsize=(16,7))
    sns.distplot(df[col], hist=hist, kde=True, 
             bins=bins, color = color, 
             hist_kws={'edgecolor':'black'},
            kde_kws={'linewidth': 3},
                label =label)
    if show :
        if label != None:
            plt.legend()
        plt.title(title)
        plt.show()
    
```

```python
def normalize(data):
    mean = np.mean(data,axis =0)
    return (data - mean)/np.std(data,axis = 0 )
data = credit_cards.drop(columns = ["wealth","expenditure_class"]).to_numpy(dtype = "float64")
normalized_data = normalize(data)
```

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
def covariance_matrix(matrix,N):
    return np.divide(np.cov(matrix.T),N)
def sort_eigs(eigenvalues, eigenvectors):
    pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]
    pairs.sort(reverse=True)
    eigvalues_sort  = [pairs[index][0] for index in range(len(eigenvalues))]
    eigvectors_sort = [pairs[index][1] for index in range(len(eigenvalues))]
    return np.array(eigvalues_sort),np.array(eigvectors_sort)
    
N,_ = normalized_data.shape
cov_matrix = covariance_matrix(normalized_data,N)
eigenvalues, eigenvectors= np.linalg.eig(cov_matrix)
eig_val,eig_vector =sort_eigs(eigenvalues, eigenvectors) 
print(f'eigenvalues:\n{eig_val}\n')
print(f'eigenvectors:\n{eig_vector}\n')
```
